## MatPlotLib

```python
import matplotlib.pyplot as plt
x=np.arange(2,10)
y=np.arange(12,20)
print(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
plt.plot(x,y,color='red',label='Peso')
plt.legend()
plt.grid()
plt.show()
```

```python
z=[25,12,11,14,13,10,17,3]
plt.plot(x,z,color='blue',label='Masa')
plt.plot(x,y,color='red',label='Peso')
#plt.title('Grfica')
plt.xlabel('Eje X')
plt.ylabel('Eje y')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.bar(y,x ,edgecolor='black',color='red')

plt.show()
```

```python
plt.pie(x,labels=x)
plt.show
```

```python
plt.scatter(z,x)
plt.show()
```

```python
#tomado de : https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')


plt.show()
```
