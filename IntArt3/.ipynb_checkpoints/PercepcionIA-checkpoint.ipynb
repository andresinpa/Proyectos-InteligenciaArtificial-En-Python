<br><h4>3.1 Escala de grises üî≤</h4>
<p>üëâ Se utiliza la funci√≥n rgb2gray del subm√≥dulo color de la biblioteca scikit-image para convertir una imagen en color en una imagen en escala de grises</p>
<br>

```python
# El argumento img[...,0:3] se utiliza para extraer los canales de color rojo, verde y azul de la imagen. 
img_gris=color.rgb2gray(img[...,0:3])
io.imshow(img_gris) #Se puede posteriormente usar io.show() para esconder el AxesImage
io.show() 
```

<br><p>üëâ Se importa el m√≥dulo pyplot de la biblioteca matplotlib y lo asigna al alias plt, esto nos serve para ajustar las visualizaciones de las imagenes, posteriormente en una sola impresi√≥n se muestra la imagen original vs la de escala de grises</p><br>

```python
# Se crea una figura con dos subgr√°ficos para cada imagen
plt.subplot(221) #subplot(matriz 2x2 en el indice 1)
plt.title('Original')
io.imshow(img)
plt.subplot(222) #subplot(matriz 2x2 en el indice 2)
plt.title('Escala de grises')
io.imshow(img_gris)
io.show()
```

<br>
<h4>3.2 Binarizaci√≥n de imagen ‚óª‚óº </h4>
<p>üëâ Se aplica nuevamente una escala de grises a la imagen para posterirmente acceder a cada pixel desde el recorrido de filas y columnas, dependiendo si es mayor o menor a 0.5 se asignan valores (0 o 1) para cambiar el pixel y binarizar la imagen.</p><br>

```python
img_gray = color.rgb2gray(img) #Conversi√≥n de imagen a grises
rows,cols = img_gray.shape #dimensiones de la imagen

#recorrido y binarizaci√≥n
for i in range(rows):
    for j in range(cols):
        if (img_gray[i,j] <= 0.5):
            img_gray[i,j]= 0.0
        else:
            img_gray[i,j]= 1.0

io.imshow(img_gray)
io.show()
        
```

<br><h4>3.3 Ruido Gaussiano üîä</h4>
<p>üëâ Se importa la libreria skimage para posteriormente hacer uso del submodulo util y random_noise o ruido gaussiano que es  un tipo de ruido que se distribuye normalmente y puede ser generado mediante una distribuci√≥n de probabilidad gaussiana.</p>
<br>

```python
import skimage
```

```python
#Ruido Gaussiano
#... .random_noise(imgen original, tipo de ruido a aplicar, varianza del ruido gaussiano)()
noise=skimage.util.random_noise(img, mode='gaussian', var=0.15)
io.imshow(noise)
```

<br><p>üëâSe importa la librer√≠a numpy y ahora se agrega ruido gaussiano a la imagen utilizando la funci√≥n np.random.normal de NumPy.</p><br>

```python
import numpy as np
```

```python
mean = 0.5  #Media
var = 0.1 #Varianza

noise = np.random.normal (mean, var**0.05, img.shape)
#Se multiplica la imagen original por la matriz de ruido noise para agregar el ruido a la imagen
noisy = img * noise
#se limitan los valores de la matriz resultante noisy entre 0 y 1 
noisy = np.clip(noisy,0.0,1.0)
io.imshow(noisy)
```

<br><h3>4. Scipy üëÄ</h3>  
<h4>Procesesamiento de imagenes a partir de pixeles</h4>
<p>üëâ misc es un subm√≥dulo de scipy que contiene diversas funciones para trabajar con im√°genes, como por ejemplo para leer y escribir im√°genes, operaciones de manipulaci√≥n de im√°genes y transformaciones geom√©tricas.
ndimage es otro subm√≥dulo de scipy que proporciona una gran cantidad de funciones para operaciones de procesamiento de im√°genes, como filtrado, segmentaci√≥n, morfolog√≠a matem√°tica, entre otras.</p>
<br>

```python
#Importando submodulos de la librer√≠a scipy
from scipy import misc  
from scipy import ndimage
```

```python
#Procesamiento de imagen, genera ruido en la imagen a partir de una escala de grises
#1.5*img_gris.std() controla la amplitud del ruido
noisy = img_gris + 1.5*img_gris.std()*np.random.random(img_gris.shape)
io.imshow(noisy)
```

<br>
<p>üëâ En la l√≠nea gauss_denoised = ndimage.gaussian_filter(noisy, 2.8), se est√° aplicando un filtro gaussiano a la imagen noisy con un valor de desviaci√≥n est√°ndar de 2.8. Esto suaviza la imagen y reduce el ruido generado anteriormente.</p>
<br>

```python
#Reduciendo Ruido
gauss_denoised = ndimage.gaussian_filter(noisy, 2.8) 
io.imshow(gauss_denoised)
```

<br>
<p>üëâ Filtros a la imagen en escala de grises</p>
<br>

```python
#aplicando un filtro gaussiano con un valor de desviaci√≥n est√°ndar de 2. Esto suaviza la imagen y reduce el ruido.
blurred_img = ndimage.gaussian_filter(img_gris, sigma=2)
#filtro gaussiano con un valor de desviaci√≥n est√°ndar de 8. Esto suaviza a√∫n m√°s la imagen y reduce m√°s el ruido.
very_blurred = ndimage.gaussian_filter(img_gris, sigma=8)
#se est√° aplicando un filtro uniforme con un tama√±o de ventana de 2x2 p√≠xeles. 
#Cada p√≠xel se reemplaza por el promedio de los p√≠xeles de su vecindario de 2x2
local_mean = ndimage.uniform_filter(img_gris, size=2)
```

<br>
<p>üëâ Importando la libreria matplotlib y el submodulo pyplot cada imagen se muestra en una subfigura separada en el mismo espacio de visualizaci√≥n.</p>
<br>

```python
import matplotlib.pyplot as plt
```

```python
# 2x2, posicion 1
plt.subplot(221)
plt.title('Original')
io.imshow(img)

# 2x2, posicion 2
plt.subplot(222)
plt.title('blurred_img')
io.imshow(blurred_img)

# 2x2, posicion 2
plt.subplot(223)
plt.title('very_blurred')
io.imshow(very_blurred)

# 2x2, posicion 4
plt.subplot(224)
plt.title('local_mean')
io.imshow(local_mean)
io.show()
```

<br>
<p>üëâ El siguiente c√≥digo utiliza una t√©cnica de procesamiento de im√°genes para detectar bordes en una imagen mediante la diferenciaci√≥n de la intensidad de los p√≠xeles.</p>
<br>

```python
#Aplica un filtro gaussiano a la imagen para suavizarla.
im = ndimage.gaussian_filter(img, 3.5)
#Sobel es un operador de gradiente utilizado para detectar bordes en im√°genes.
#Se aplica sobel en direcciones horizontal (axis=0) y vertical (axis=1) a la imagen suavizada
sx = ndimage.sobel(im, axis=0, mode='constant')
sy = ndimage.sobel(im, axis=1, mode='constant')
#se calcula la magnitud de los gradientes obtenidos en cada direcci√≥n
sob = np.hypot(sx, sy)

io.imshow(sy)
```

<br><h4>5. Detecci√≥n de bordes</h4>
<p>üëâ El siguiente c√≥digo utiliza el submodulo filters de la librer√≠a Scikit-image para realizar la detecci√≥n de bordes en una imagen en escala de grises. Se aplican diferentes filtros de detecci√≥n de bordes (Sobel, Scharr y Prewitt)</p>
<br>

```python
from skimage import filters

#creaci√≥n de tres variables donde se llaman cada uno de los filtros a usar
filtros = [filters.sobel, filters.roberts, filters.prewitt]

#Procesamiento para la deteccion de bordes
edge_sobel = filters.sobel(img)
edge_sobel = filters.sobel(img_gris)
edge_scharr = filters.scharr(img_gris)
edge_prewitt = filters.prewitt(img_gris)

#Se genera una figura con 3 subfiguras en una sola fila (ncols=3)
fig, axes = plt.subplots(ncols=3, sharex=True, sharey=True,
                         figsize=(18, 18))

#comparacion con escala de grises
axes[0].imshow(edge_sobel, cmap=plt.cm.gray)
axes[0].set_title('sobel Edge Detection')

axes[1].imshow(edge_scharr, cmap=plt.cm.gray)
axes[1].set_title('scharr Edge Detection')

axes[2].imshow(edge_prewitt, cmap=plt.cm.gray)
axes[2].set_title('prewitt Edge Detection')
```

```python
edge_roberts = filters.roberts(img_gris)
edge_sobel = filters.sobel(img)

fig, axes = plt.subplots(ncols=2, sharex=True, sharey=True,
                         figsize=(15, 6))

axes[0].imshow(edge_roberts, cmap=plt.cm.gray)
axes[0].set_title('Roberts Edge Detection')

axes[1].imshow(edge_sobel, cmap=plt.cm.gray)
axes[1].set_title('Sobel Edge Detection')

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
